```python
def education(x):
    if x=='Secondary / secondary special':
        x=x.split(' / ')[0]
    return x

dataset['NAME_EDUCATION_TYPE']=dataset['NAME_EDUCATION_TYPE'].apply(education)
```

```python
def fun(x):
    if x=='Single / not married':
        x=x.split(' / ')[0]
    return x

dataset['NAME_FAMILY_STATUS']=dataset['NAME_FAMILY_STATUS'].apply(fun)

dataset['NAME_FAMILY_STATUS'].value_counts()
```

```python
## This function takes no of days and convert it into their datetime format
def Date_of_Birth(total_days):
    today = datetime.date.today()
    birthday = (today + timedelta(days=total_days)).strftime('%Y-%m-%d')
    return birthday                  

## This Function takes value of colunn Days of Employed and convert it into datetime format
def Employed_day(total_days):
    today = datetime.date.today()
    employed_date = (today + datetime.timedelta(days=total_days)).strftime('%Y-%m-%d')
    
dataset['DAYS_BIRTH']=dataset['DAYS_BIRTH'].apply(Date_of_Birth)
dataset['DAYS_EMPLOYED']=dataset['DAYS_EMPLOYED'].apply(Date_of_Birth)
```

```python
def AGE(date_of_birth):
    today_year= 2019  # The dataset represents 2019's. 
    dateofbirth=datetime.datetime.strptime(date_of_birth,'%Y-%m-%d')
    age=today_year-dateofbirth.year
    return age

dataset['DAYS_BIRTH']=dataset['DAYS_BIRTH'].apply(AGE)
```

```python
def solve(state):
    target=''
    if state in (2,3,4,5):
        target=1 #Risky
    else :
        target=0 #not Risky
        
    return target
```
